```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv("aac_intakes.csv")
data.head()
```

```python
animal_type = pd.DataFrame(data.animal_type.value_counts().reset_index().values, columns=["Animal Type", "Number of Animals"])
animal_type = animal_type.sort_index(axis = 0, ascending=True)
animal_type
```

```python
plt.figure(figsize = (9, 6)) #determine the size of the chart
# To create a bar chart with matplotlib you can use the 'bar' function.
plt.bar(x = animal_type["Animal Type"], #specify the x axis
        height = animal_type["Number of Animals"], #specify the y axis
        color = "m") # the color for the bars
plt.xticks(rotation = 45, fontsize = 13) # rotate and format the labels for the x-axis
plt.yticks(fontsize = 13) # format the y-axis
plt.title("Number of Animal Intake", fontsize = 16, fontweight = "bold") #add and format the title for the chart
plt.ylabel("Number of Animals", fontsize = 13 ) #add a title for the y-axis
plt.savefig("Number of Animals Intakes.png") # you can export your chart as a picture
plt.show() # depending on which environment you're using without this line you're chart may not display properly.
```

```python
intake_condition = pd.DataFrame(data.intake_condition.value_counts().reset_index().values, columns=["Intake Condition", "Number of Animals"])
intake_condition = intake_condition.sort_index(axis = 0, ascending=True)
intake_condition
```

```python
plt.figure(figsize = (9, 6)) 
plt.bar(x = intake_condition["Intake Condition"], 
        height = intake_condition["Number of Animals"], 
        color = "m") 
plt.xticks(rotation = 45, fontsize = 13) 
plt.yticks(fontsize = 13) 
plt.title("Intake Condition of Animals", fontsize = 16, fontweight = "bold") 
plt.ylabel("Number of Animals", fontsize = 13 ) 
plt.savefig("Intake Condition of Animals.png") 
plt.show() 
```
